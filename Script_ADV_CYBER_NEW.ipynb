## PARSER

```python
import os
import csv
import re
from random import sample
import math

# PARAMS
PATH = os.path.join(path_progettoADV, '_log_files')
OUT_FILENAME = 'out_temp_file_10_perc.csv'
LEN_THRESHOLD = 100
DNs_PERCENTAGE = 10

# get the list of files
files_list = os.listdir(PATH)
# open output file
out_file = open(os.path.join(path_progettoADV, OUT_FILENAME), 'w', newline='')
writer = csv.writer(out_file)
# domain names list
DNs_list = []
DNs_list_samples = []
# for each file
count_over_100 = 0
count_no_regex = 0
records = 0

for filename in files_list:
    file = open(os.path.join(PATH, filename), 'r')
    records = records + len(file.readlines())
print('number of total records : {}'.format(records))

for filename in files_list:
    # skip not '.log' files
    if '.log' not in filename:
      continue
    # open file
    file = open(os.path.join(PATH, filename), 'r')
    # for each line
    for i, line in enumerate(file):
        # get domain name
        DN_name = line.split(';')[5]

        # clear final dot in the domain name
        if DN_name[-1] == '.':
            DN_name = DN_name[0:-1]
        # skip domain names with length greater than threshold
        # TODO: aggiunta condizione <3 altrimenti creava problemi nella creazione degli n-grams
        
        if len(DN_name) > LEN_THRESHOLD or len(DN_name) < 3:
            count_over_100 = count_over_100 + 1
            continue

        # skip domain names with addresses
        if re.match(r'.*\d*\.\d*\.\d*', DN_name):
          count_no_regex = count_no_regex + 1
          continue  
        else:
          DNs_list.append(DN_name)
    # random choice 10% of DNs
    DNs_list_sample = sample(DNs_list, math.floor((DNs_PERCENTAGE*len(DNs_list))/100))
    DNs_list_samples += DNs_list_sample
# get unique domain names
set_of_DNs = set(DNs_list_samples)
print(f'Estratti {len(set_of_DNs)} nomi di dominio.')
# write domain names and domain names without dots
writer.writerows([value, f'{value}'.replace('.', '')] for value in set_of_DNs)

print('record over 100, less 3 : {}'.format(count_over_100))
print('record no respect regex : {}'.format(count_no_regex))
#tempo di esecuzione circa 4m
```

## Creazione 1-grams, 2-grams e 3-grams

```python
import csv
from nltk import ngrams

# PARAMS
IN_FILENAME = 'out_temp_file_10_perc.csv'
OUT_FILENAME = 'out_final_file_10_perc.csv'

# open input and output files
with open(os.path.join(path_progettoADV, IN_FILENAME), 'r') as in_file, open(os.path.join(path_progettoADV, OUT_FILENAME), 'w', newline='') as out_file:
    reader = csv.reader(in_file)
    writer = csv.writer(out_file)
    for row in reader:
        # get domain name without dots
        domain_name = row[1]
        
        # extract 1-grams from the domain name
        n = 1
        one_grams = ngrams(domain_name, n)
        one_gram = [''.join(el) for el in one_grams]

        # extract 2-grams from the domain name
        n = 2
        two_grams = ngrams(domain_name, n)
        two_gram = [''.join(el) for el in two_grams]

        # extract 3-grams from the domain name
        n = 3
        three_grams = ngrams(domain_name, n)
        three_gram = [''.join(el) for el in three_grams]

        # write extracted info to the output file
        writer.writerow([row[0], row[1], ' '.join(one_gram), ' '.join(two_gram), ' '.join(three_gram)])
#tempo di esecuzione circa 
```

### UTILS per CM from CSV to Confusion Matrix (miglioramento visivo)

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import math 

def get_cm_image (path=None, filename=None): 
  os.chdir(path)
  #read csv e etl label 
  CSVFILE = filename
  df_label = pd.read_csv(CSVFILE)
  print('DF NO ETL SHAPE : {}'.format(df_label.shape))
  df_etl = pd.read_csv(CSVFILE, header=0, usecols=[*range(1,52)]) 
  print('DF ETL SHAPE : {}'.format(df_etl.shape))

  #normalize 0.01
  scaler = MinMaxScaler()
  scaler.fit(df_etl)
  scaled = scaler.fit_transform(df_etl)
  scaled_df=pd.DataFrame(scaled,columns=df_etl.columns)
  trunc = lambda x: math.trunc(10 * x) / 10;
  scaled_df = scaled_df.applymap(trunc)

  #plot fig
  fig, figa = plt.subplots(figsize=(30, 30), dpi=180)
  ax = sns.heatmap(scaled_df, annot=True, cmap='Blues', fmt="g", ax=figa)

  ax.set_title('Confusion Matrix\n\n');
  ax.set_xlabel('\nPredicted Values')
  ax.set_ylabel('Actual Values ');

  ## Ticket labels 
  ax.xaxis.set_ticklabels(df_label.iloc[:,0], rotation = 45, ha="right")
  ax.yaxis.set_ticklabels(df_label.iloc[:,0], rotation = 45, ha="right")

  ## Display
  #plt.show()
  #SAVE FIG 
  
  plt.savefig(dpi=180, figsize=(30, 30), transparent=False, fname='CM-FINAL.png')
```
